```python
Fibonacci Series

#include <iostream>
using namespace std;

int fibo(int n) {
    if (n == 0 || n == 1) {
        return n;
    } else {
        return fibo(n - 1) + fibo(n - 2);
    }
}

int main() {
    int n;

    cout << "Enter a number: ";
    cin >> n;

    cout << "Fibonacci Series upto " << n << " is:\n";

    for (int i = 0; i < n; i++) {
        cout << fibo(i) << " ";
    }

    return 0;
}

Binary Search (recursive)
#include <iostream>
using namespace std;

int binarySearchRecursive(int arr[], int low, int high, int key) {
    if (low <= high) {
        int mid = (low + high) / 2;

        if (arr[mid] == key)
            return mid;
        else if (arr[mid] > key)
            return binarySearchRecursive(arr, low, mid - 1, key);
        else
            return binarySearchRecursive(arr, mid + 1, high, key);
    }
    return -1;
}

int main() {
    int n, key;

    cout << "Enter number of elements: ";
    cin >> n;

    int arr[n];
    cout << "Enter sorted elements: ";
    for (int i = 0; i < n; i++)
        cin >> arr[i];

    cout << "Enter element to search: ";
    cin >> key;

    int result = binarySearchRecursive(arr, 0, n - 1, key);

    if (result != -1)
        cout << "Element found at index: " << result;
    else
        cout << "Element not found";

    return 0;
}


Binary Search (Non - recursive)

#include <iostream>
using namespace std;

int binarySearchIterative(int arr[], int n, int key) {
    int low = 0, high = n - 1;

    while (low <= high) {
        int mid = (low + high) / 2;

        if (arr[mid] == key)
            return mid;
        else if (arr[mid] < key)
            low = mid + 1;
        else
            high = mid - 1;
    }
    return -1;
}

int main() {
    int n, key;

    cout << "Enter number of elements: ";
    cin >> n;

    int arr[n];
    cout << "Enter sorted elements: ";
    for (int i = 0; i < n; i++)
        cin >> arr[i];

    cout << "Enter element to search: ";
    cin >> key;

    int result = binarySearchIterative(arr, n, key);

    if (result != -1)
        cout << "Element found at index: " << result;
    else
        cout << "Element not found";

    return 0;
}


Heap Sort

#include <iostream>
using namespace std;

// Heapify function (Max Heap)
void heapify(int arr[], int n, int i) {
    int largest = i;
    int left = 2*i + 1;
    int right = 2*i + 2;

    if (left < n && arr[left] > arr[largest])
        largest = left;

    if (right < n && arr[right] > arr[largest])
        largest = right;

    if (largest != i) {
        swap(arr[i], arr[largest]);
        heapify(arr, n, largest);
    }
}

// Heap Sort function
void heapSort(int arr[], int n) {

    // Build max heap
    for (int i = n/2 - 1; i >= 0; i--)
        heapify(arr, n, i);

    // Extract elements
    for (int i = n-1; i > 0; i--) {
        swap(arr[0], arr[i]);
        heapify(arr, i, 0);
    }
}

int main() {
    int n;

    cout << "Enter number of elements: ";
    cin >> n;

    int arr[n];

    cout << "Enter elements:\n";
    for (int i = 0; i < n; i++) {
        cin >> arr[i];
    }

    heapSort(arr, n);

    cout << "Sorted array: ";
    for (int i = 0; i < n; i++) {
        cout << arr[i] << " ";
    }

    return 0;
}


Knapsack Code:
#include<bits/stdc++.h>
using namespace std;

struct Item {
    int profit;
    int weight;
};

    bool static pwRatio(Item a, Item b) {
        double r1 = (double) a.profit / a.weight;
        double r2 = (double) b.profit / b.weight;
        return r1 > r2;
    }
    
    double fractionalKnapsack(int sackCapacity, Item arr[], int n){
        sort(arr, arr + n, pwRatio);
        
        int curWeight = 0;
        double finalProfit = 0.0;
        
        for(int i = 0; i < n; i++){
            
            if(curWeight + arr[i].weight <= sackCapacity){
                curWeight += arr[i].weight;
                finalProfit += arr[i].profit;
            } else {
                int remCapacity = sackCapacity - curWeight;
                finalProfit += ((double)remCapacity/(double)arr[i].weight) * arr[i].profit;
                break;
            }
        }
        
        return finalProfit;
    }


int main() {
    int n, weight = 0;
    cout << "\nEnter the number of items (n): ";
    cin >> n;
    Item arr[n] = {{0,0}};
    
    for(int i = 0; i < n; i++){
    cout << "\nEnter profit for item " << i+1 << ": "; 
        cin >>  arr[i].profit;
    cout << "Enter weight of item " << i+1 << " : ";
        cin >>  arr[i].weight;
        
    }
    
    cout << "Enter the capacity of Knapsack (W): ";
    cin >> weight;
    double ans = fractionalKnapsack(weight, arr, n);
    
    cout << "The maximum profit is: " << setprecision(2) << fixed << ans;
    
    return 0;
}


Job Sequencing 
#include <iostream>
#include <algorithm>
using namespace std;

struct Job {
    char id;
    int deadline;
    int profit;
};

bool comparison (Job a, Job b){
    return (a.profit > b.profit);
}

void JobScheduling(Job arr[], int n){

    sort(arr, arr + n, comparison);

    int result[n]; 
    bool slot[n];
    
    for (int i = 0; i < n; i++)
    slot[i] = false;


    for(int i = 0; i < n; i++){
        for(int j = min(n,arr[i].deadline) - 1; j>=0; j--){
            if(slot[j] == false){
                slot[j] = true;
                result[j] = i;
                break;
            }
        }
    }

    for(int i = 0; i < n; i++){
        if(slot[i]){
            cout << arr[result[i]].id;
        }
    }
}

int main()
{
int n;
cout << "Enter number of jobs: ";
cin >> n;
Job arr[n];
for (int i = 0; i < n; i++)
{
cout << "Enter Job ID, Deadline, Profit: ";
cin >> arr[i].id >> arr[i].deadline >> arr[i].profit;
}
cout << "\nMaximum profit job sequence:\n";
JobScheduling(arr, n);
return 0;
}


Bellman Ford

#include <iostream>
#include <climits>
using namespace std;

struct Edge {
    int u, v, w;
};

int main() {
    int V = 5; // number of vertices
    int E = 8; // number of edges

    Edge edges[8] = {
        {0, 1, -1},
        {0, 2, 4},
        {1, 2, 3},
        {1, 3, 2},
        {1, 4, 2},
        {3, 2, 5},
        {3, 1, 1},
        {4, 3, -3}
    };

    int dist[5];
    int source = 0;

    // Initialize distances
    for (int i = 0; i < V; i++) {
        dist[i] = INT_MAX;
    }

    dist[source] = 0;

    // Relax edges V-1 times
    for (int i = 1; i <= V - 1; i++) {
        for (int j = 0; j < E; j++) {
            if (dist[edges[j].u] != INT_MAX &&
                dist[edges[j].u] + edges[j].w < dist[edges[j].v]) {

                dist[edges[j].v] = dist[edges[j].u] + edges[j].w;
            }
        }
    }

    // Check for negative weight cycle
    for (int j = 0; j < E; j++) {
        if (dist[edges[j].u] != INT_MAX &&
            dist[edges[j].u] + edges[j].w < dist[edges[j].v]) {

            cout << "Graph contains negative weight cycle\n";
            return 0;
        }
    }

    // Print result
    cout << "Vertex\tDistance from Source\n";
    for (int i = 0; i < V; i++) {
        cout << i << "\t" << dist[i] << endl;
    }

    return 0;
}


N-Queen

#include <iostream>
using namespace std;

// Function to print the board
void printBoard(int board[][10], int n) {
    for (int i = 0; i < n; i++) {
        for (int j = 0; j < n; j++) {
            cout << board[i][j] << " ";
        }
        cout << endl;
    }
    cout << endl;
}

// Function to check if a queen can be placed at (row, col)
bool isSafe(int board[][10], int row, int col, int n) {

    // Check column
    for (int i = 0; i < row; i++)
        if (board[i][col])
            return false;

    // Check upper-left diagonal
    for (int i = row - 1, j = col - 1; i >= 0 && j >= 0; i--, j--)
        if (board[i][j])
            return false;

    // Check upper-right diagonal
    for (int i = row - 1, j = col + 1; i >= 0 && j < n; i--, j++)
        if (board[i][j])
            return false;

    return true;
}

// Backtracking function
bool solveNQueens(int board[][10], int row, int n) {

    // Base case: all queens placed
    if (row == n) {
        printBoard(board, n);
        return true;
    }

    bool foundSolution = false;

    for (int col = 0; col < n; col++) {
        if (isSafe(board, row, col, n)) {

            board[row][col] = 1; // place queen

            foundSolution = solveNQueens(board, row + 1, n) || foundSolution;

            board[row][col] = 0; // backtrack
        }
    }

    return foundSolution;
}

int main() {
    int n = 4;  // change value of N here

    int board[10][10] = {0};

    if (!solveNQueens(board, 0, n)) {
        cout << "No solution exists" << endl;
    }

    return 0;
}


Quick Sort
#include <iostream>
using namespace std;

int partition(int arr[], int l, int r, int pivot)
{
    while (l <= r)
    {
        while (arr[l] < pivot)
        {
            l++;
        }

        while (arr[r] > pivot)
        {
            r--;
        }

        if (l <= r)
        {
            swap(arr[l], arr[r]);
            l++;
            r--;
        }
    }
    return l;
}

void quick_sort(int arr[], int l, int r)
{
    if (l < r)
    {
        int mid = l + (r - l) / 2;
        int pivot = arr[mid];

        int index = partition(arr, l, r, pivot);

        quick_sort(arr, l, index - 1);
        quick_sort(arr, index, r);
    }
}

int main()
{
    int n;
    cout << "Enter number of elements: ";
    cin >> n;

    int arr[50];

    cout << "Enter elements:\n";
    for (int i = 0; i < n; i++)
    {
        cin >> arr[i];
    }

    quick_sort(arr, 0, n - 1);

    cout << "Sorted array:\n";
    for (int i = 0; i < n; i++)
    {
        cout << arr[i] << " ";
    }

    return 0;
}

```
